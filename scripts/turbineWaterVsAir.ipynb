Comparing the design of a water turbine to wind turbine energy harvesting system.

The power generated by a wind turbine is:

$$P_\text{wind} = \frac{16}{27}\frac{1}{2} \rho_\text{a} v^3_\text{a} A_\text{wind} \eta_\text{wind} $$

where the subscript "a" denotes air and that generated by a water turbine is:

$$P_\text{water} = \frac{16}{27}\frac{1}{2} \rho_\text{w} v^3_\text{w} A_\text{water} \eta_\text{water} $$

To understand the performance of a wind vs. water turbine, we can take the ratio of the above power generation equations, we have:

$$\frac{P_\text{wind}}{P_\text{water}} = \frac{\frac{16}{27}\frac{1}{2} \rho_\text{a} v^3_\text{a} A_\text{wind} \eta_\text{wind}}{\frac{16}{27}\frac{1}{2} \rho_\text{w} v^3_\text{w} A_\text{water} \eta_\text{water}} $$

If the areas, efficiencies, and velocities are the same, we have:

$$\frac{P_\text{wind}}{P_\text{water}} = \frac{\rho_\text{a} }{\rho_\text{w} } $$
