## Specific heats of a monatomic gas

The internal energy of an ideal monatomic gas was derived previously as:

$$U = \frac{3}{2}k_B N T $$

Recalling:

$$R = k_B N_A $$

we can write:

$$U = \frac{3}{2}n R T $$

When the temperature changes from $T$ to $T + dT$, the change in internal energy is:

$$dU = \frac{3}{2}n R dT $$

If the process is constant volume, so that no work is done, using the first law we know:

$$\Delta U = Q $$

That is, when no work is done, the change in internal energy of the substance is equal to the heat absorbed. We can therefore write:

$$\left(dQ\right)_{V=\text{const}} = dU = \frac{3}{2} n R dT  $$

$$C = \frac{1}{n}\frac{dQ}{dT} $$

The heat absorbed depends on the process. Therefore, there is a heat capacity for each process. The two most common are heat capacity at constant pressure, $C_p$, and the heat capacity at constant volume, $C_V$:

$$\bar{c}_p = \frac{1}{n} \left(\frac{dQ}{dT}\right)_p $$

$$\bar{c}_V = \frac{1}{n} \left(\frac{dQ}{dT}\right)_V $$

Therefore, we have:

$$\bar{c}_V = \frac{1}{n} \left( \frac{3}{2} n R \right) = \frac{3}{2} R $$

We can then write the internal energy as:

$$dU = n \bar{c}_V dT $$

If the change in temperature is at constant pressure, the work done by the gas is:

$$W = \int_{V_1}^{V_2}p dV = p\int_{V_1}^{V_2} dV = p \left(V_2 - V_1 \right) $$

From ideal gas equation, we can write

$$pV = k_B N T $$

Substituting above yields (const pressure):

$$ W = k_B N\left(T_2 - T_1 \right) = n R \left(T_2 - T_1 \right)$$

From the first law we have:

$$ Q = \Delta U + W$$

So the heat absorbed in a constant pressure process is:

$$\left( dQ \right)_{p = \text{const}} = dU + dW$$

$$\left( dQ \right)_{p = \text{const}} = \frac{3}{2} n R dT  + n R dT = \frac{5}{2} n R dT $$

Dividing both sides by $dT$ yields:

$$\left( \frac{dQ}{dT} \right)_{p = \text{const}} = \frac{5}{2} n R$$

$$\bar{c}_p = \left( \frac{dQ}{dT} \right)_{p = \text{const}} = \frac{5}{2} n R $$

The ratio of $\bar{c}_p/\bar{c}_V$ is:

$$\gamma = \bar{c}_p/\bar{c}_V = \frac{5}{3} $$

and

$$R = \bar{c}_p - \bar{c}_V $$

## Statistical mechanics

Recall that the translational partition function, which we derived, is:

$$ Z_{tr} = \left( \frac{2 \pi m k T}{h^2}\right)^{3/2} V $$

$$ \ln Z_{tr} = \frac{3}{2}\ln \left( \frac{2 \pi m k}{h^2}\right) + \frac{3}{2}\ln \left( T \right) + \ln( V) $$

$$ \left(\frac{\partial \ln Z_{tr}}{\partial T}\right)_V = \frac{3}{2T}$$

we can now evaluate the contributions of the translational mode to thermodynamic properties of an ideal gas:

$$\left( \frac{u}{R T}\right) = T \left(\frac{\partial \ln Z_{tr}}{\partial T}\right)_V$$

Substituting the above:

$$\left( \frac{u}{R T}\right) = T \left(\frac{3}{2T}\right) = \frac{3}{2}$$

The specific heat at constant volume is:

$$ \left(\frac{\bar{c}_V}{R}\right) = \left[\frac{\partial}{\partial T}T^2\left( \frac{\partial \ln Z_{tr}}{\partial T} \right)\right]_V = \frac{3}{2}$$

## Adiabatic Transformation of an Ideal Gas

When a system undergoes an adiabatic transformation, from the first law we have:

$$\Delta U = - W $$

or

$$dU = - dW $$

When the work is due to a change in volume, 

$$dW = p dV $$ 

and

$$dU = - p dV $$

For an ideal gas we can write

$$n \bar{c}_VdT + pdV = 0 $$

We can write the ideal gas equation in differential form:

$$ p dV + V dp = n R dT $$

$$dT = -\frac{pdV}{n \bar{c}_V} $$

$$ p dV + V dp = n R \left( -\frac{pdV}{n \bar{c}_V} \right) $$

$$ \bar{c}_Vp dV + \bar{c}_VV dp + R \left( \frac{pdV}{1} \right)= 0 $$

$$ \left(\bar{c}_V + R \right)p dV  + \bar{c}_VV dp = 0 $$

Recalling:

$$R = \bar{c}_p - \bar{c}_V $$

We can write:

$$ \bar{c}_p p dV  + \bar{c}_VV dp = 0 $$

From the definition:

$$\gamma = \frac{\bar{c}_p}{\bar{c}_V} $$

We can write:

$$ \frac{\bar{c}_p}{\bar{c}_V} p dV  + V dp = 0 $$

$$ \gamma p dV  + V dp = 0 $$

$$ \frac{dp}{p} + \gamma  \frac{dV}{V}   =  0$$

Integrating:

$$ \int \frac{dp}{p} + \gamma \int \frac{dV}{V}   =  \text{const}$$

$$\ln p + \gamma \ln V = \text{const}$$

$$ \implies p V^{\gamma} = \text{const}$$
