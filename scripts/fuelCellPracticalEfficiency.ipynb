Let us consider a hydrogel fuel cell. The reaction is:

$$H_2(g) + (1/2)O_2(g) \rightarrow H_2O(g) $$

The rate at which water vapor is generated is $\bar{\dot{n}}_\text{p,H2O}$. The number of electrons produced per molecule of the product is $n_e$. The rate that electrons are delivered by the cell to the load is:

$$n_e N_0 \bar{\dot{n}}_\text{p,H2O}$$

The current is then:

$$I = q n_e N_0 \bar{\dot{n}}_\text{p,H2O} $$

For hydrogen/oxygen cells, $n_e = 2$.

Recall that the theoretical efficiency of a reversible fuel cell is:

$$\eta_\text{rev} = \frac{\Delta G}{\Delta H} $$

The efficiency of practical fuel cells is the ratio of the electrical power, $P_\text{L}$, delivered to the load to the heat power that would be generated by combining the reactants under the same temperature and pressure used in the cell:

$$\eta_\text{practical} = \frac{P_\text{L}}{P_\text{in}} = \frac{I_\text{L} V_\text{L}}{\Delta \bar{h}_\text{comb} \bar{\dot{n}}_\text{p,H2O}}$$

Substituting the equation for current, assuming 100% current efficiency ($I_L = I$), yields:

$$\eta_\text{practical} = \frac{P_\text{L}}{P_\text{in}} = \frac{(q n_e N_0 \bar{\dot{n}}_\text{p,H2O}) V_\text{L}}{\Delta \bar{h}_\text{comb} \bar{\dot{n}}_\text{p,H2O}} = \frac{q n_e N_0  V_\text{L}}{\Delta \bar{h}_\text{comb} }$$
