```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
I_sc = 5.0  # Short-circuit current (A)
V_oc = 0.6  # Open-circuit voltage (V)
n = 1.5      # Diode ideality factor
V_t = 0.025  # Thermal voltage (V)

# Define voltage range
V = np.linspace(0, V_oc, 100)

# Shockley diode equation-based approximation
I = I_sc * (1 - np.exp((V - V_oc) / (n * V_t)))

# Plot I-V curve
plt.figure(figsize=(8, 6))
plt.plot(V, I, label='I-V Curve', color='blue')
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel('Voltage (V)')
plt.ylabel('Current (A)')
plt.savefig('I-V-curve.png')
plt.show()
```
