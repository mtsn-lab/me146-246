### Fermi-Dirac Distribution

Electrons in a metal do not follow a Maxwell-Boltzmann distribution. They are governed by the Pauli exclusion principle, which means that no two electrons can have the same quantum numbers in a system. Therefore, they cannot all have zero energy at absolute zero. At absolute zero, electrons are distributed uniformly in energy up to a given energy level. No electron has energy above that level, termed the Fermi level. The probability, $f$, of finding electrons at a given energy level is:

$$f = 1 \text{, for } W < W_F \text{ and } f = 0 \text{ for } W > W_F$$

The number of allowed states with momenta less than $|p|$ is proportional to the volume of a sphere with radius $|p|$ in the momentum space:

$$ N \text{ (with momenta less than |p|)} = \frac{2}{h^3} \left(\frac{4}{3}\pi p^3\right)$$

The proportionality constant is $2/h^3$ where 2 is the result of two possible spins of the electron and $N$ is the number of electrons per cubic meter. 

Since $p = mv$ it follows that $p = \sqrt{2mW}$ and

$$ N \text{ (with momenta less than |p|)} = \frac{2}{h^3} \left(\frac{4}{3}\pi \left(\sqrt{2mW}\right)^3\right)$$

The partial derivative 

$$\frac{\partial N}{\partial W} $$ 

is then:

$$\frac{\partial N}{\partial W} = \frac{8 \sqrt{2} \pi}{h^3}m^{3/2}W^{1/2} $$

When the temperature is greater than zero, the probability that a given state is occupied is given by:

$$f = \frac{1}{1 + \exp\left(\frac{W - \mu}{k_B T}\right)} $$

where $\mu$ is the chemical potential. When $T = 0$, $\mu = W_F$ and the function above has the property that $f=1$ for $W<W_F$ and $f=0$ for $W>W_F$ as required.

The density of states using the probability function becomes:

$$\frac{\partial N}{\partial W} = \frac{8 \sqrt{2} \pi}{h^3}m^{3/2} \frac{W^{1/2}}{1 + \exp\left(\frac{W - \mu}{k_B T}\right)} $$

and integrating yields:

$$N = \frac{8 \sqrt{2} \pi}{h^3}m^{3/2} \int_0^\infty \frac{W^{1/2}}{1 + \exp\left(\frac{W - \mu}{k_B T}\right)}dW $$



The chemical potential is defined as:


$$\mu = W_F - \frac{(\pi k_B T)^2}{12 W_F} $$

For temperatures less than 2000 K, one can take $\mu = W_F$. 

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure()
plt.clf()
W = np.linspace(0, 2, 100) # range of temperatures

mu_eV = 5
mu_J = 1.60218e-19*mu_eV

h = 1
m = 1
mu = 1
k = 1
T = 0.5

dNdW = [((8*(2**(1/2))*3.1415)/(h**3))*(m**(3/2))*((w**(1/2))/(1+np.exp((w-mu)/(k*T)))) for w in W]

plt.plot(W, dNdW, 'k-')

plt.xlabel('W (energy))')
plt.ylabel('dN/dW (-)')

# plt.savefig('hydrogen-T-dG.png')

#Tsat2 = np.linspace(200, 400, 1000) # range of temperatures
#plt.plot(svap, Tsat2, 'r-')
#plt.plot(sliq, Tsat2, 'r-')
```
