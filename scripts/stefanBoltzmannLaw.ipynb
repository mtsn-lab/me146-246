### Kinetic theory of gases with classical particles

It can be shown that the relationship between the gas pressure, $p$, and the square of the arithetic mean gas particle velocity, $\bar{v}^2$, is:

$$p = \frac{1}{3} \frac{n}{V}m \bar{v}^2 $$

where $n$ is the total number of particles, $V$ is volume, and $m$ is the mass of a single particle.

For ideal gases, the kinetic energy of the particles is $(1/2)m\bar{v}^2 $. Therefore, we can see that the above equation can be re-written as:
$$p = \frac{2}{3} \frac{n}{V} \bar{w}_\text{kin} $$

The total internal energy of the gas is then $n \bar{w}_\text{kin}$, which yields:
$$p = \frac{2}{3} \frac{U}{V}  $$

and can be written as energy density per volume, $u_V= U/V$:
$$p = \frac{2}{3} u_V  $$

### Kinetic theory of gases with relativistic particles

We can write the pressure of a photon gas as:

$$p = \frac{1}{3}\frac{N}{V}m_\text{rel} c^2 $$

where $c$ is the speed of light (all photons travel with the same speed). Recall also:

$$E = m c^2 $$ and $$W = m_\text{rel} c^2 $$

where $m_\text{rel}$ is the relativistic mass (not rest mass). We can determine the relativistic mass from quantum mechanics:

$$W = h f $$

where $h$ is the Planck constant and $f$ is the light frequency. We can re-write our pressure equation as:

$$p = \frac{1}{3}\frac{N}{V} w $$

$$p = \frac{1}{3}\frac{U}{V} $$

$$p = \frac{1}{3} u_V $$

Compare with the classical case above. 

### Thermodynamics

Start with the first law. We assume that a photon gas is analogous to a classical ideal gas. Therefore:

$$dU = TdS - p dV $$

Dividing by $dV$, and assuming an isothermal process, we can write:

$$\left( \frac{\partial U}{\partial V} \right)_T = T\left( \frac{\partial S}{\partial V}\right)_T - p $$

Using Maxwell's relations, we can write (see Thermodynamic square):

$$\left( \frac{\partial S}{\partial V} \right)_T = \left(\frac{\partial p}{\partial T} \right)_V$$

Substituting into the above equation we see:

$$\left( \frac{\partial U}{\partial V} \right)_T = T\left(\frac{\partial p}{\partial T} \right)_V - p $$

We can deal with the term on the left-hand side by recalling that the energy of radiation from a blackbody only depends on temperature. Therefore, for constant temperature, and as a result, constant energy density, $u_V$, increasing the volume by $dV$ leads to an increase of the photon gas energy $dU$:

$$U(T,V) = u_V(T)V $$

$$\frac{dU}{dV} = \frac{d(u_V(T) V)}{dV} $$

$$\frac{dU}{dV} = \frac{d(u_V(T))}{dV}V + u_V\frac{d( V)}{dV} $$

$$\frac{dU}{dV} = u_V $$

From our equation for pressure above, we then have:

$$\frac{dU}{dV} = u_V = 3p $$

Coming back to our formula for the first law, we then see that the left-hand side is replaced by $u_V$:

$$u_V = T\left(\frac{\partial p}{\partial T} \right)_V - p $$

Substituting $p = u_V/3$ we see that:

$$u_V = T\left(\frac{\partial u_V/3}{\partial T} \right)_V - u_V/3 $$

Re-arranging yields:
$$3u_V = T\left(\frac{\partial u_V}{\partial T} \right)_V - u_V $$

Since $u_V$ only depends on temperature, the partial derivative becomes an ordinary derivative:

$$4u_V = T \frac{d u_V}{d T}$$

Rearranging yields:

$$4 \frac{dT}{T} =  \frac{d u_V}{u_V}$$

Integrating yields:

$$\int 4 \frac{dT}{T} = \int \frac{d u_V}{u_V}$$

$$ 4 \ln T + a = \ln u_V$$

where $a$ is a constant of integration. Rearranging yields:

$$u_V = a T^4 $$

### Radiant power

For photons emitted from a surface area, $A$, travelling at the speed of light, $c$, they will occupy a volume $dV$ above the surface after a time $dt$:

$$dV = A c dt $$

The change in internal energy is then:

$$dU = u_V dV = a T^4 A c dt $$

The radiant power, $\Phi$, is then the change in internal energy with respect to $t$:

$$\Phi = \frac{dU}{dt} = a c A T^4  $$

where $ \sigma = a c $ and is called the Stefan-Boltzmann constant (5.67 x 10$^{-8}$ W/m$^2$/K$^4$). For a blackbody, radiant power is related to intensity, $I$, as:

$$I = \Phi / A = \sigma T^4$$

For a real body with an emissivity $\epsilon < 1$, the intensity is:

$$I = \epsilon \sigma T^4$$
